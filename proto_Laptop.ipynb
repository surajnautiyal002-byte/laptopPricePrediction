```python
import re
import pandas as pd

def clean_memory(mem):
    try:
        mem = str(mem).replace('Flash Storage', 'Flash')
        mem = mem.replace(' ', '').upper()

        ssd = hdd = hybrid = flash = 0

        parts = mem.split('+')

        for part in parts:
            size_match = re.findall(r'\d+\.?\d*', part)
            if not size_match:
                continue

            size = float(size_match[0])
            if 'TB' in part:
                size *= 1024  # Convert TB to GB

            if 'SSD' in part:
                ssd += size
            elif 'HDD' in part:
                hdd += size
            elif 'HYBRID' in part:
                hybrid += size
            elif 'FLASH' in part:
                flash += size

        return pd.Series([ssd, hdd, hybrid, flash])
    except:
        return pd.Series([0, 0, 0, 0])

# Apply the function directly to the column
df[['SSD', 'HDD', 'Hybrid', 'Flash']] = df['Memory'].apply(clean_memory)

```

```python
df['TotalMemory'] = df['SSD'] + df['HDD'] + df['Hybrid'] + df['Flash']

```

```python
import pandas as pd
import numpy as np
import re

# Function to convert TB and GB to GB
def convert_to_gb(text):
    num = float(re.search(r'\d+\.?\d*', text).group())
    if 'TB' in text:
        return int(num * 1024)
    elif 'GB' in text:
        return int(num)
    return 0

# Initialize columns
df['SSD'] = 0
df['HDD'] = 0
df['FlashStorage'] = 0
df['Hybrid'] = 0

# Fill values based on 'Memory' column
for i, row in df.iterrows():
    mem_str = row['Memory']
    for part in mem_str.split('+'):
        part = part.strip()
        if 'SSD' in part and 'Flash' not in part:
            df.at[i, 'SSD'] += convert_to_gb(part)
        elif 'HDD' in part:
            df.at[i, 'HDD'] += convert_to_gb(part)
        elif 'Flash Storage' in part:
            df.at[i, 'FlashStorage'] += convert_to_gb(part)
        elif 'Hybrid' in part:
            df.at[i, 'Hybrid'] += convert_to_gb(part)

# Drop original column if needed
# df.drop(columns=['Memory'], inplace=True)
```

```python
import pandas as pd
import re

# Mapping of CPU keywords to type
cpu_type_map = {
    'intel core i9': 'i9',
    'intel core i7': 'i7',
    'intel core i5': 'i5',
    'intel core i3': 'i3',
    'intel pentium': 'pentium',
    'intel celeron': 'celeron',
    'intel xeon': 'xeon',
    'amd ryzen 7': 'ryzen7',
    'amd ryzen 5': 'ryzen5',
    'amd ryzen 3': 'ryzen3',
    'amd a12': 'a12',
    'amd a9': 'a9',
    'amd a6': 'a6',
}

# CPU performance tiers (numerical)
cpu_tier_map = {
    'i9': 3,
    'ryzen7': 3,
    'xeon': 3,
    'i7': 2,
    'ryzen5': 2,
    'i5': 2,
    'ryzen3': 2,
    'a12': 2,
    'i3': 1,
    'a6': 1,
    'pentium': 1,
    'celeron': 0,
    'a9': 0,
    'other': 1  # default
}

def extract_cpu_features(cpu_str):
    cpu_lower = cpu_str.lower()
    cpu_type = 'other'
    for key in cpu_type_map:
        if key in cpu_lower:
            cpu_type = cpu_type_map[key]
            break

    # Extract GHz value
    match = re.search(r'(\d+\.?\d*)ghz', cpu_lower)
    ghz = float(match.group(1)) if match else None

    # Assign tier
    tier = cpu_tier_map.get(cpu_type, 1)
    return pd.Series([cpu_type, ghz, tier])

# Apply to DataFrame
df[['CpuType', 'CpuSpeedGHz', 'CpuTier']] = df['Cpu'].apply(extract_cpu_features)

```

```python
def simplify_os(os):
    if 'Windows' in os:
        return 'Windows'
    elif 'Mac' in os or os == 'macOS':
        return 'Mac'
    elif 'Linux' in os:
        return 'Linux'
    elif 'Chrome' in os:
        return 'Chrome OS'
    elif 'Android' in os:
        return 'Android'
    else:
        return 'Other'

df['OpSys_Simplified'] = df['OpSys'].apply(simplify_os)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['OpSys_Code'] = le.fit_transform(df['OpSys_Simplified'])
```
